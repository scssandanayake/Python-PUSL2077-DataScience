```python


df['year'] = pd.DatetimeIndex(df['date']).year
```

```python
columns_to_encode = ['new_cases']

label_encoder = LabelEncoder()

for column in columns_to_encode:
    df[column] = label_encoder.fit_transform(df[column])
```

```python
import numpy as np
import matplotlib.pyplot as plt
w = 0.2
x = ['USA','Russia','GBR','China','Finland']
gold = [20,10,6,15,30]
silver = [100,50,30,120,200]
bronze = [200,100,60,240,300]

# bar1 = np.arange(len(x)) - 0.2
# bar2 = np.arange(len(gold)) + 0.4
# bar3 = np.arange(len(silver)) +0.2

bar1 = np.arange(len(x))
bar2 = [i +w for i in bar1]
bar3 = [i +w for i in bar2]

plt.bar(bar1,gold,0.4,label = 'gold')
plt.bar(bar2,silver,0.4,label = 'silver')
plt.bar(bar3,bronze,0.4,label = 'bronze')

plt.xlabel('Countries')
plt.xticks(bar1+w,x)
plt.ylabel('medals')
plt.title('medals vs countries')
plt.legend()
```
