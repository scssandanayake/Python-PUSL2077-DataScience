```python
# import libraries
import numpy as np 
import pandas as pd 
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.preprocessing import StandardScaler
from sklearn.decomposition import PCA
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import classification_report, confusion_matrix, ConfusionMatrixDisplay
import warnings
warnings.filterwarnings('ignore')
```

```python
# display options
pd.set_option('display.max_columns', None)
```

```python
# read dataset
df = pd.read_csv('Loan_Default.csv')
```

```python
# first five rows
df.head()
```

```python
# dataset shape
df.shape
```

```python
# dataset info
df.info()
```

```python
# change dtype of Status and term
df = df.astype({'term':'object', 'Status':'object'})
```

```python
# null values
df.isnull().sum()
```

```python
# plot null value count
plt.figure(figsize=(12, 6))
df.isnull().sum().plot.bar()
plt.title('Null Value Count By Column')
plt.xlabel('Column')
plt.ylabel('Null value count')
plt.xticks(rotation = 90)
plt.show()
```

```python
# remove ID and year 
df_1 = df.drop(['ID','year'], axis=1)
df_1.head()
```

```python
# checking for unique values
df_1.nunique().sum
```

```python
# separate categorical and numerical columns
categorical_cols = []
numerical_cols = []
for col in df_1.columns:
    if(df_1[col].dtype == 'object'):
        categorical_cols.append(col)
    else:
        numerical_cols.append(col)
```

```python
print('Categorical columns: \n', categorical_cols)
print('\n')
print('Numerical columns: \n', numerical_cols)
```

```python
# imputing missing values of numerical columns with means
df_1_num = df_1[numerical_cols]
num_means = df_1_num.mean()
```

```python
df_1_num.fillna(num_means, inplace=True)
```

```python
df_1_num.isnull().sum()
```

```python
# checking for unique values in categorical columns to find any inconsistencies
df_1_cat = df_1[categorical_cols]
for col in categorical_cols:
    print(df_1_cat[col].value_counts(), "\n")    
```

```python
# imputing missing values of categorical columns with modes
for col in categorical_cols:
    cat_mode = df_1_cat[col].mode()
    cat_mode = cat_mode[0]
    df_1_cat[col].fillna(value=cat_mode, inplace=True)
```

```python
df_1_cat = df_1_cat.astype({'term':'object'})
```

```python
df_1_cat.isnull().sum()
```

```python
# concatenating two dataframes
df_2 = pd.concat([df_1_num, df_1_cat], axis=1, join='inner')
```

```python
df_2.isnull().sum()
```

```python
df_2.info()
```

```python
# outlier detection
df_2.plot(kind="box",subplots=True,layout=(5,2),figsize=(20,20))
```

```python
df_2.shape
```

```python
df_3 = df_2.copy()
```

```python
# removing outliers
for col in numerical_cols:
    iqr = 1.5 * (np.percentile(df_3[col], 75) - np.percentile(df_3[col], 25))
    upper_bound = iqr + np.percentile(df_3[col], 75)
    lower_bound = np.percentile(df_3[col], 25) - iqr
    df_3.drop(df_3[(df_3[col] > upper_bound) | (df_3[col] < lower_bound)].index, inplace=True)
```

```python
df_3.shape
```

```python
# after removing
df_3.plot(kind="box",subplots=True,layout=(5,2),figsize=(20,20))
```

```python
# violin plots for numerical columns vs Status
for col in numerical_cols:
    plt.figure(figsize=(5, 5))
    sns.violinplot(x='Status', y=col, data=df_3)
    plt.title(f'Status vs {col}')
    plt.xlabel('Status')
    plt.ylabel(col)
    plt.show()
```

```python
# change dtype of Status from object to int64
df_3 = df_3.astype({'Status':'int64'})
```

```python
df_3.info()
```

```python
categorical_cols.remove('Status')
numerical_cols.append('Status')
```

```python
# preprocessing  - encoding and dropping categorical columns
df_3['is_cf'] = df_3['loan_limit'].replace({'cf': 1, 'ncf': 0})

ohe_gender = df_3['Gender'].str.get_dummies()
df_3 = pd.concat([df_3, ohe_gender], axis=1)

df_3['is_nopre'] = df_3['approv_in_adv'].replace({'nopre': 1, 'pre': 0})

ohe_loan_type = df_3['loan_type'].str.get_dummies()
df_3 = pd.concat([df_3, ohe_loan_type], axis=1)

ohe_loan_purpose = df_3['loan_purpose'].str.get_dummies()
df_3 = pd.concat([df_3, ohe_loan_purpose], axis=1)

df_3['is_nob/c'] = df_3['business_or_commercial'].replace({'nob/c': 1, 'b/c': 0})

df_3['is_360.0'] = df_3['term'].map(lambda x: 1 if x == 360.0 else 0)

df_3['is_not_neg'] = df_3['Neg_ammortization'].replace({'not_neg': 1, 'neg_amm': 0})

df_3['is_not_lpsm'] = df_3['lump_sum_payment'].replace({'not_lpsm': 1, 'lpsm': 0})

df_3['is_pr'] = df_3['occupancy_type'].map(lambda x: 1 if x == 'pr' else 0)

ohe_credit_type = df_3['credit_type'].str.get_dummies()
df_3 = pd.concat([df_3, ohe_credit_type], axis=1)

df_3['is_exp'] = df_3['co-applicant_credit_type'].replace({'EXP': 1, 'CIB': 0})

ohe_age = df_3['age'].str.get_dummies()
df_3 = pd.concat([df_3, ohe_age], axis=1)

df_3['is_to_inst'] = df_3['submission_of_application'].replace({'to_inst': 1, 'not_inst': 0})

ohe_region = df_3['Region'].str.get_dummies()
df_3 = pd.concat([df_3, ohe_region], axis=1)

df_4 = df_3.drop(categorical_cols, axis=1)
```

```python
df_4.info()
```

```python
# correlation analysis
corr = df_4.corr()
plt.figure(figsize=(30, 30))
sns.heatmap(data = corr, vmin=-1, vmax=1, annot=True)
```

```python
# remove highly inter correlated columns
df_5 = df_4.drop(['North', 'south', 'is_nob/c', 'type2', 'loan_amount', 'property_value', ], axis=1)
```

```python
# split X and y
X = df_5.drop(['Status'], axis=1)
y = df_5['Status']
```

```python
X.shape
```

```python
y.shape
```

```python
# scaling and PCA for feature selection
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)
pca = PCA(n_components=10)
X_pca = pca.fit_transform(X_scaled)
```

```python
# split the dataset into train and test sets
X_train, X_test, y_train, y_test = train_test_split(X_pca, y, stratify=y, test_size=0.25)
```

```python
# model building and training
lr_model = LogisticRegression()
lr_model.fit(X_train, y_train)
```

```python
# model predictions
lr_pred = lr_model.predict(X_test)
lr_score = lr_model.score(X_test, y_test)
print(lr_score)
```

```python
# classification report
print(classification_report(y_test, lr_pred))
```

```python
# confusion matrix
cm = confusion_matrix(y_test, lr_pred)
disp = ConfusionMatrixDisplay(confusion_matrix=cm)
disp.plot()
plt.show()
```
