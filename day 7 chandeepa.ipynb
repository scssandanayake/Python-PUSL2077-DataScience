#### Two ways to create a matplotlib plot

#### 1)Functional Appraoch

```python
plt.plot(sldc['time line'],sldc['total_cases'])
```

#### Object oriented approach

```python
fig = plt.figure() #create a figure conatining a single axes
ax = fig.add_axes([0,0,3,1]) #left,bottom,width,height
ax.plot(sldc['time line'],sldc['total_cases'])
```

```python
import matplotlib.pyplot as plt

fig, ax = plt.subplots()
ax.plot(sldc['time line'], sldc['total_cases'], label='total_cases')
ax.set_xlabel('Time Line')
ax.set_ylabel('Total Cases')
ax.set_title('Covid cumulative')
ax.legend(loc='best')

plt.show()
```

### Visualize the subplots

```python

x = sldc['time line']
y = ['total_cases', 'new_cases', 'total_deaths', 'new_deaths']

# Create subplots
fig, a = plt.subplots(2, 2, sharex='col')  # sharex='row' if you prefer

# Plot each data on its corresponding subplot
a[0][0].plot(x, sldc[y[0]])
a[0][0].set_title(y[0])

a[0][1].plot(x, sldc[y[1]])
a[0][1].set_title(y[1])

a[1][0].plot(x, sldc[y[2]])
a[1][0].set_title(y[2])

a[1][1].plot(x, sldc[y[3]])
a[1][1].set_title(y[3])

# Adjust layout
plt.tight_layout()

# Show plot
plt.show()
```

```python
x = sldc['time line']
y = ['cardiovasc_death_rate', 'diabetes_prevalence', 'female_smokers', 'male_smokers']

fig, a = plt.subplots(2, 2, sharex='col')

a[0][0].plot(x, sldc[y[0]])
a[0][0].set_title(y[0])

a[0][1].plot(x, sldc[y[1]])
a[0][1].set_title(y[1])

a[1][0].plot(x, sldc[y[2]])
a[1][0].set_title(y[2])

a[1][1].plot(x, sldc[y[3]])
a[1][1].set_title(y[3])

plt.tight_layout()
plt.show()
```

## Insets

##### Plot insoide a plot

```python
from mpl_toolkits.axes_grid1.inset_locator import inset_axes
```

```python
fig, (ax, ax2) = plt.subplots(1, 2, figsize=[3*2, 3])  # axes sizes are given under figsize
fig.tight_layout(pad=2)
axins = inset_axes(ax, width=1.3, height=0.9)
axins2 = inset_axes(ax2, width="30%", height="40%", loc=4)
ax.tick_params(labelleft=False, labelbottom=False)
```
