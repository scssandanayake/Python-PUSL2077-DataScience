### object oriented approach

2.object oriented interface: this is the best way to create plots.the concept is create figure objects and call methods off it.creating a blank figure using the .figure() method

```python
fig = plt.figure() #creeate a figure containing a single axis
ax =fig.add_axes([0,0,3,1]) #left,bottom,width,height
ax.plot(sicd['time line'],sicd['total_cases'])
```
